```python
# Step 1 Import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# from numpy import mean,median
# load dataset
kashti = sns.load_dataset("titanic")
kashti
```

```python
# Bar plot graph with color and order
sns.barplot(x="sex",y="alone",hue='who',data=kashti,order=["female","male"],color="red")
plt.show()
```

```python
# Bar plot graph with errorbar and palette
sns.barplot(x="sex",y="alone",hue='who',data=kashti,order=["female","male"],color="red",errorbar=None,palette='pastel'
)
plt.show()
```

```python
# Bar plot graph with errorbar and palette
sns.barplot(x="class",y="fare",hue='sex',data=kashti,estimator=median)
plt.show()
```

```python
sns.barplot(x="class",y="fare",hue='sex',data=kashti,estimator=median,saturation=0.1)
plt.show()
```

```python
# Horizantal plot
sns.barplot(x="fare",y="class",hue='sex',data=kashti,estimator=median,saturation=1)
plt.show()
```

```python
sns.barplot(x="class",y="fare",data=kashti,linewidth=5,facecolor=(0.1,0.3,0.3,0.5),errcolor=".2",edgecolor=".2")
plt.show()
```

```python
# canvas (Balloon board)
sns.set(style="whitegrid")
sns.boxplot(x="class",y="fare",data=kashti)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
kashti = sns.load_dataset("titanic")
sns.boxplot(x="survived",y="age",showmeans=True,meanprops={"marker":"+",
"markersize":"12","markeredgecolor":"red"},
data=kashti)
plt.show()
```

```python
sns.boxplot(x="survived",y="age",showmeans=True,meanprops={"marker":"+",
"markersize":"12","markeredgecolor":"red"},data=kashti)
plt.xlabel("How many survived",size=10),
plt.ylabel("Age(years)",size=10),
plt.title("Boxplots",size=10)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")

# Load the example diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Draw a scatter plot while assigning point colors and sizes to different
# variables in the dataset
f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.despine(f, left=True, bottom=True)
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.scatterplot(x="carat", y="price",
                hue="clarity", size="depth",
                palette="ch:r=-.2,d=.3_r",
                hue_order=clarity_ranking,
                sizes=(1, 8), linewidth=0,
                data=diamonds, ax=ax)
```
