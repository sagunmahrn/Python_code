# EDA
Exploratory Data Analysis (EDA) is an approach to analyze the data using visual techniques. It is used to discover trends, patterns, or to check assumptions with the help of statistical summary and graphical representations. 

```python
# import libraries
import pandas as pd
import  numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load dataset
df = sns.load_dataset('titanic')
```

```python
df.columns
```

```python
# visualize the nan value in heatmap
sns.heatmap(df.isnull(),yticklabels=False, cbar=False ,cmap='viridis')
```

```python
# remove deck column
df.drop(columns="deck",inplace=True)
```

```python
df.columns # or use df.shape
```

```python
# visualize and explore
sns.countplot(data=df,x="survived")
```

```python
sns.barplot(data=df,x="survived",y="age",hue="who")
```

```python
sns.countplot(x="pclass",data=df)
```

```python
# data distribution
sns.histplot(data=df,x="age",bins=10)
```

```python
df['age'].hist(bins=30,color='darkred',alpha=0.75)
```

```python
sns.boxplot(data=df,y="age", x="survived",hue="who")
```

```python
#scatterplot
sns.scatterplot(data=df,x='age',y='fare',hue='pclass')
plt.ylim=(0,300)
```

```python
sns.pairplot(df,hue='survived')
```

```python
# stacked barchart/plot
df.groupby(['pclass','who'])['fare'].mean().unstack().plot(kind='bar',stacked=True)
```

```python
#swarmplot
sns.swarmplot(data=df,x='survived',y='age',hue='who')
```

```python
sns.catplot(x='pclass',y='fare',hue='embarked',kind='point',data=df)
```

```python
# relation of two numeric variables

sns.lmplot(data=df[df['fare']<300],x='age',y='fare')
```

```python
# grouping
df.columns
```

```python
df['age_groups'] = pd.cut(df['age'],[0,10,18,30,50,80],labels=['Kids','Child','Youth','Middle aged','Old'])
```

```python
df.head()
```

```python
df.columns
```

```python
df.groupby('age_groups')['survived'].mean().plot(kind="bar")
```

```python
sns.pairplot(df,hue='survived',vars=['age','fare','pclass'])
```
