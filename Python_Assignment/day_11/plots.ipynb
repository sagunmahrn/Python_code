# Plotting for Python

```python
# import libararies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import dataset
kashti=sns.load_dataset('titanic')
kashti
```

```python
kashti.info()
```

```python
kashti.isnull().sum()
```

```python
kashti.survived.value_counts().plot(kind="bar")
```

```python
sns.countplot(x="survived",hue="sex",data=kashti)
```

```python
#scatter plot
plt.scatter(data=kashti,x="age",y="fare",color="pclass")
```

```python
sns.scatterplot(data=kashti,x="age",y="fare",hue="pclass")
```

```python
sns.barplot(data=kashti,x="sex",y="fare",hue="pclass")
```

```python
kashti.columns
```

```python
sns.pairplot(kashti)
```

```python
kashti.hist(bins=10,figsize=(9,6),gid=False)
```

```python
kashti.columns
```

```python
sns.boxplot(data=kashti,x="sex",y="fare",hue="class")
```

```python
sns.boxenplot(data=kashti,x="sex",y="fare",hue="class")
```

```python
# Assignment:how to change the x and y axs title?
ax=sns.boxplot(data=kashti,x="sex",y="fare",hue="class")
ax.set(xlabel='Sex', ylabel='Fare')
```

```python
g =sns.FacetGrid(kashti,hue="survived",col="who",margin_titles=True,
palette={1:"seagreen",0:"gray"})
g=g.map(plt.scatter,"fare","age",edgecolor="w").add_legend();
```

```python
sns.stripplot(data=kashti,x="sex",y="fare",hue="class")
```

```python
# heat map with correlation
corr =kashti.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr)
plt.title("Heat map of Kashti Data")
```

```python
sns.violinplot(data=kashti,x="sex",y="fare",hue="class")
```
